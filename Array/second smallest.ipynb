```python
def secondsmallest():
    arr=[1,2,4,7,7,5]
    smallest = arr[0]
    ss = float('inf')
    for i in range (1,len(arr)):
        if arr[i] < smallest:
            ss = smallest
            smallest = arr[i]
        elif arr[i] != smallest and arr[i]<ss:
            ss = arr[i]
        return (ss)
secondsmallest()
```
