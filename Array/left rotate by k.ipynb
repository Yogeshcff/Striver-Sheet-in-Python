```python
def leftrotateb_by_k(k):
    
    arr=[1,2,3,4,5,6,7]
    k=k%len(arr)
    temp=arr[0:3]
    for i in range (k,len(arr)):
        arr[i-k]=arr[i]
    for j in range (len(arr)-k,len(arr)):
        arr[j]=temp[j+k-len(arr)]
    return arr
leftrotateb_by_k(7)

```

```python
def optimal(d):
    arr=[1,2,3,4,5,6,7]
    arr[0:d]=arr[0:d][::-1]
    arr[d:len(arr)]=arr[d:len(arr)][::-1]
    arr[0:len(arr)]=arr[0:len(arr)][::-1]
    return arr
optimal(3)
```
