```python
class solution:
    def pattern(self,n):
        
        for a in range (1,n+1):
            for b in range (a,n+1):
                print("*",end=" ")
            for c in range (a,a+1):
                print("  "*(a*2-2),end="")
            for d in range (n+1,a,-1):
                print("*",end=" ")
            print()
        x=n
        for i in range (1,n+1):
            for j in range (1,i+1):
                print("*",end=' ')
            for k in range (0,2*x-2):
                print(" ",end=" ")
            for l in range (i,0,-1):
                print("*",end=" ")
            x=x-1
            
            print()
       
a=solution()
a.pattern(5)
```

```python
class solution:
    def pattern(self,n):
        
        for i in range (0,2*n):
            for j in range (0,2*n):
                if i==0 or i==2*n-1 or j==0 or j==2*n-1:
                    print("4",end=" ")
                elif i==1 or j==1 or i==2*n-2 or j==2*n-2:
                    print("3",end=" ")
                elif i==2 or j==2 or j==2*n-3 or i==2*n-3:
                    print("2",end=" ")
                else:
                    print(" ",end=" ")
            print()
       
a=solution()
a.pattern(5)
```
